2) IMPORTAÇÕES DE BIBLIOTECAS

```python
import os
import pytesseract
from PIL import Image, UnidentifiedImageError
from pdf2image import convert_from_path

# Para upload e download de arquivos no Colab
from google.colab import files
```

3) FUNÇÃO PARA EXTRAIR TEXTO

```python
def extrair_texto_arquivo(caminho: str, idioma: str = 'por') -> str:
    """
    Extrai texto de um arquivo (imagem ou PDF).
    Se for imagem, faz OCR diretamente.
    Se for PDF, converte para imagens e então faz OCR.

    :param caminho: Caminho para o arquivo (imagem ou PDF).
    :param idioma: Idioma de reconhecimento (por padrão, 'por').
    :return: Texto extraído do arquivo.
    """
    try:
        # 1) Tenta abrir como imagem
        with Image.open(caminho) as img:
            return pytesseract.image_to_string(img, lang=idioma)
    except UnidentifiedImageError:
        # 2) Se falhar, tenta converter PDF para imagens e faz OCR em cada página
        paginas_imagens = convert_from_path(caminho, dpi=300)
        texto_total = ""
        for i, imagem in enumerate(paginas_imagens, start=1):
            texto_pagina = pytesseract.image_to_string(imagem, lang=idioma)
            texto_total += f"\n=== [Página {i}] ===\n"
            texto_total += texto_pagina
        return texto_total
```

4) CARREGAMENTO/USO DA FUNÇÃO

```python
print("Faça upload de um arquivo (imagem ou PDF) para o Colab...")
uploaded = files.upload()

if uploaded:
    # Pegamos o primeiro nome de arquivo enviado
    nome_arquivo = list(uploaded.keys())[0]

    if os.path.exists(nome_arquivo):
        texto_extraido = extrair_texto_arquivo(nome_arquivo, idioma='por')

        # Cria o arquivo texto_extraido.txt
        nome_arquivo_txt = "texto_extraido.txt"
        with open(nome_arquivo_txt, "w", encoding="utf-8") as f:
            f.write(texto_extraido)

        print(f"\nO texto foi extraído e salvo em: {nome_arquivo_txt}")

        # (Opcional) Baixar o arquivo gerado
        # files.download(nome_arquivo_txt)
    else:
        print(f"O arquivo {nome_arquivo} não foi encontrado no diretório atual.")
else:
    print("Nenhum arquivo foi enviado.")
```
